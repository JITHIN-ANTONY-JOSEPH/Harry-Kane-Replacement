# Level the Playing Field

```python
epl_fw.head(2)
```

```python
convert_dict = {'id': int
                ,'player_name'   : object
                ,'games'   : int
                ,'time'   : int
                ,'goals'   : int
                ,'xG'   : float
                ,'assists'   : int
                ,'xA'   : float
                ,'shots'   : int
                ,'key_passes'   : int
                ,'yellow_cards'   : int
                ,'red_cards'   : int
                ,'position'   : object
                ,'team_title'   : object
                ,'npg'   : int
                ,'npxG'   : float
                ,'xGChain'   : float
                ,'xGBuildup'   : float
               }
```

```python
epl_fw = epl_fw.astype(convert_dict)
print(epl_fw.dtypes)
```

```python
epl_fw[epl_fw.games>10]
```

```python
epl_fw=epl_fw.sample(  frac=1)
```

```python
epl_fw.shape
```

```python
fig2 = px.bar(epl_fw, x='player_name', y='goals')
fig2.show()
```

```python
# Since we might have many columns to analyse per 90 we'll create a loop to calcualte for all
per90Cols=['goals','xG','assists','xA',
           'shots','key_passes','npg','npxG','xGChain','xGBuildup']

for col in per90Cols:
    epl_fw[col+"Per90"]=epl_fw[col].divide(epl_fw['time']).multiply(90)      
```

```python
epl_fw.shape
```

```python
epl_fw.head(10)
```

```python
# Factoring in descipline into consideration

# Since we might have many columns to analyse per 90 we'll create a loop to calcualte for all

for col in per90Cols:
    epl_fw['warnings'+"Per90"]=epl_fw['yellow_cards'].divide(epl_fw['time']).multiply(90)     
```

```python

cols_for_radar=[i+'Per90' for i in per90Cols]
```

```python
cols_for_radar.append('warningsPer90')
```

```python
cols_for_radar
```

```python
# Checking the Plot
px.line_polar(epl_fw,r=epl_fw[cols_for_radar][epl_fw.id==1250].sum(),
              theta=cols_for_radar,line_close=True)
```

```python
epl_fw[cols_for_radar]=scaler.fit_transform(epl_fw[cols_for_radar])
```

```python
for i,row in epl_fw.iterrows():
    print(row['player_name'])
    fig=px.line_polar(epl_fw,r=epl_fw[cols_for_radar][(epl_fw.id==row.id)].sum(),
                theta=cols_for_radar,line_close=True,title=row['player_name'])
    fig.update_traces(fill='toself')
    fig.show()
```

```python
isl_max=epl_fw[cols_for_radar].max().max()
```

```python
import plotly.graph_objects as go
```

```python
for i,row in epl_fw.iterrows():
    if row.id==647:
        continue
    if(row.goalsPer90>=0.4):
        print(row['player_name'])
        fig4=go.Figure()
        fig4.add_trace(go.Scatterpolar(r=epl_fw[cols_for_radar][epl_fw.id==647].sum(),
                    theta=cols_for_radar,fill='toself',name='Harry Kane'))
        fig4.add_trace(go.Scatterpolar(r=epl_fw[cols_for_radar][(epl_fw.id==row['id'])].sum(),
                    theta=cols_for_radar,fill='toself',name=row['player_name']))

        # Adding additional properties
        fig4.update_layout(
            title="Harry Kane vs " + row["player_name"],
            polar=dict(
                    radialaxis=dict(
                        #visisble=True,
                        range=[0,isl_max]
                        )),
            showlegend=True
            )   
        fig4.show()
```
